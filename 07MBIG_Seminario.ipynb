```
#BoxPlot de cada provincia
with sns.axes_style(style='ticks'):
  ax = sns.catplot("Provincia", "Tasa_PromocionEfectiva", data=promocion, kind="box", height=10, aspect=2)
  ax.set_xticklabels(rotation=30)
```
